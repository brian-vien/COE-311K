```python
import numpy as np
import matplotlib.pyplot as plt
import math

def taylor(x, nterms):
    result = 0
    for n in range(nterms):
        term = ((-1)**n)*(x**(2*n+1))/math.factorial(2*n+1)
        result += term
    return result

xvalues = np.arange(0,10.1,0.1)

precisevalues = np.sin(xvalues)

error = {}

for terms in [1,2,3,4]:
    approxvalues = np.array([taylor(x,terms) for x in xvalues])
    abserror = np.abs(approxvalues - precisevalues)
    error[terms] = abserror
    
plt.figure()
for terms in [1,2,3,4]:
    plt.plot(xvalues, error[terms], label = f"{terms}terms")
    
plt.xlabel("x")
plt.ylabel("absolute error")
plt.title("Truncation error for Taylor Series of sin(x)")
plt.legend()
plt.show()
```

As more terms are added to the Taylor Series, the error decreases. The error does not decrease uniformly for all x's. If the x value is small, the terms give small error, while big x values result in larger error values.
