```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

#Below is the Monte Carlo Pi Function
def monte_carlo(num_samples):
    inside = 0
    for _ in range(num_samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <=1:
            inside +=1
            
    return 4*inside/num_samples

    #This function generates x and y points randomly in a 1x1 square,
    #and the fraction of the points inside this square can be used to approximate the value of pi
    
#Main code body, computes error and makes the samples
samples = range(10,1001,20)
error = [0] *len(samples)
    
real_pi = round(math.pi,15)
for i in range(len(samples)):
     n = samples[i]
     estimate_pi = monte_carlo(n)
     error[i] = abs(estimate_pi - real_pi)
        
#plotting the results
plt.figure()
plt.plot(samples, error, marker = 'o')
plt.xlabel("number of samples")
plt.ylabel("Absolute value of error")
plt.title("Approximation of Pi using the Monte Carlo vs Samples")
plt.show()
```
