```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_mat_multiplication(A, B):
    n = len(A)
    result = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    
    return result

sizes = []
current_size = 3

for _ in range(8):
    sizes.append(current_size)
    current_size *= 2

naive_times = []
numpy_times = []

for size in sizes:
    print(f"Processing size: {size}x{size}")
    
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)
    

    start = time.time()
    naive_mat_multiplication(A.tolist(), B.tolist())
    end = time.time()
    naive_times.append(end - start)
    
    start = time.time()
    np.dot(A, B)
    end = time.time()
    numpy_times.append(end - start)

plt.figure()

plt.plot(range(1, 9), naive_times, label='Naive O(n^3)')
plt.plot(range(1, 9), numpy_times, label='NumPy')

plt.xlabel('Matrix Number (1 to 8)')
plt.ylabel('Runtime (seconds)')
plt.title('Naive vs NumPy Matrix Multiplication Runtime')
plt.legend()

plt.show()
```

Looking at the graph, it can be seen that the numpy runtime is very fast, basically close to zero, while the naive runtime is close to zero until matrix 6, and then the runtime gets significantly longer. We can see that the numpy is much faster than the naive runtime when there are more matrices involved.
