```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import math

x = np.random.rand(500)
y = np.random.rand(500)
size = np.random.rand(500)*50
color = np.random.rand(500)
plt.subplot(3,1,1)
plt.scatter(x, y, size, color)
plt.colorbar()
plt.show()

x = np.linspace(0, 2*np.pi,100)
plt.subplot(3,1,2)
plt.plot(x, np.tan(x), 'r')
plt.show()

c = np.empty([1250,1250])
im = plt.imshow(c, cmap= 'hot')
fig = plt.colorbar(im, orientation='horizontal')
plt.show(fig)



def graph(x,y):
    return (x*x*x*x+y*y*y*y)
x = np.linspace(-10,50,1000)
y = np.linspace(-10,50,1000)
X, Y = np.meshgrid(x,y)
Z = graph(X, Y)

fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(60, 0)






```
