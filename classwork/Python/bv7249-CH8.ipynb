```python
def diagonally_dominant(matrix):
    n = len(matrix)
    
    for i in range(n):
        row_sum = sum(abs(matrix[i][j]) for j in range(n) if j != i)
        
        diagonal_element = abs(matrix[i][i])

        if diagonal_element < row_sum:
            return False
            
    return True


matrix = [
    [4, 1, 1],
    [2, 5, 1],
    [1, 2, 4]
]

print(f"Is the matrix diagonally dominant? {diagonally_dominant(matrix)}")
print()

for row in matrix:
    print(row)
    
print(f"Is the matrix diagonally dominant? {diagonally_dominant(matrix)}")
```
