```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -2 * y

def exact_sol(x):
    return np.exp(-2 * x)

x_start, x_end = 0, 10
h_values = [0.1, 0.5, 1, 1.5]

plt.figure(figsize=(12, 7))

for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_vals = np.linspace(x_start, x_end, num_steps + 1)
    y_euler = np.zeros(num_steps + 1)
    
    y_euler[0] = 1
    
    for i in range(num_steps):
        y_euler[i + 1] = y_euler[i] + h * f(x_vals[i], y_euler[i])
    
    y_exact = exact_sol(x_vals)

    local_errors = np.abs(y_exact - y_euler)
    
    global_error = local_errors[-1]
    
    print(f"Step size (h): {h:3} | Global Error at x=10: {global_error:.6f}")
    
    plt.plot(x_vals, local_errors, label=f"Error for h = {h}")

plt.xlabel('x')
plt.ylabel('Absolute Error |Exact - Euler|')
plt.title("Local Truncation Error Over Time (Euler's Method)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.yscale('log')
plt.show()
```
