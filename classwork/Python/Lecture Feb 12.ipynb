```python
import numpy as np

#creating vectors
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
print("Vector v1:", v1)
print("Vector v2:", v2)
```

```python
v_add = v1+v2
v_subtact = v1-v2
```

```python
print (v_add)
print (v_subtact)
```

```python
scalar = 2
v_multiply = scalar * v1 
print(v_multiply)
```

```python
#1*4 + 2*5 + 3*6
dot_product = np.dot(v1,v2)
print(dot_product)
```

```python
#magnitude or length of a vector (the square root of the sum of all the squared elements)

magnitude_v1 = np.linalg.norm(v1)
print(magnitude_v1)
```

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8,9],[10,11,12],[13,14,15]])
print("A:", A)
print()
print("B:",B)
```

```python
C = np.array([[1,2,3],[4,5,6]])
matrix_add = A + C
matrix_subtract = A-C
print (matrix_add)
print ()
print(matrix_subtract)
```

```python
matrix_mul = np.dot(A,B)
print(matrix_mul)
```

```python
transpose_A = A.T
print(transpose_A)
```

```python
a = np.array([[2,3],[4,-1]])
b = np.array([5,3])
print(a)
print(b)

print()
x = np.linalg.solve(a,b)
print(x)
```

```python
A = np.array([[2,1],[1,2]])
e_value, e_vectors = np.linalg.eig(A)

print("eiganvalue:", e_value)
print("eiganvector:", e_vectors)
```

```python
A = np.array([[4,7],[2,6]])
A_inv = np.linalg.inv(A)
print(A_inv)

identity= np.dot(A, A_inv)
print(identity)
```
