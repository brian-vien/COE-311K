```python
import numpy as np
import matplotlib.pyplot as plt

def euler_osc(k, h, steps=1000):
    t = np.zeros(steps)
    x = np.zeros(steps)
    v = np.zeros(steps)
    
    x[0] = 1.0 
    v[0] = 0.0  
    
    for i in range(steps - 1):
        t[i+1] = t[i] + h
        v[i+1] = v[i] - k * x[i] * h
        x[i+1] = x[i] + v[i] * h
        
    return t, x

k_val = 10
h_val = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure(figsize=(12, 7))

for h in h_val:
    total_time = 5
    num_steps = int(total_time / h)
    t, x = euler_osc(k_val, h, steps=num_steps)
    plt.plot(t, x, label=f'h = {h}')

plt.title(f'Euler Method Stability (k={k_val})')
plt.xlabel('Time (t)')
plt.ylabel('Displacement (x)')
plt.axhline(0, color='black', lw=1)
plt.legend()
plt.grid(True)
plt.show()
```
