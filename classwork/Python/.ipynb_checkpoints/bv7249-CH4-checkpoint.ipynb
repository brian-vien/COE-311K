```python
import numpy as np

approximated_pi = 0.0

for n in range(5000):
    term = ((-1)** n)/(2*n+1)
    approximated_pi += term
    
    if(n+1) % 100 ==0:
        pi_value=4*approximated_pi
        error = abs(pi_value - np.pi)/np.pi
        print ("Number of terms:", n+1, "Error:", error)
```

As the number of terms increases, the error decreases.
