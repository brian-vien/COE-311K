```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 2 * np.pi
N = 100
x = np.linspace(a, b, N)
f = np.sin

h = (b - a) / (N - 1)

df_dx = (f(x + h) - f(x - h)) / (2 * h)

forward_df = (f(x + h) - f(x)) / h

backward_df = (f(x) - f(x - h)) / h

true_df = np.cos(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, true_df, 'k', label='True cos(x)')
plt.title('True Derivative: cos(x)')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, forward_df, 'r--', label='Forward')
plt.plot(x, true_df, 'k:', alpha=0.5)
plt.title('Forward Difference')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, backward_df, 'g--', label='Backward')
plt.plot(x, true_df, 'k:', alpha=0.5)
plt.title('Backwards Difference')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, df_dx, 'b--', label='Central')
plt.plot(x, true_df, 'k:', alpha=0.5)
plt.title('Central Difference')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
