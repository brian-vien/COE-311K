```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

cs = CubicSpline(x, y, bc_type='natural')

poly_coeff = np.polyfit(x, y, 5)
poly_fun = np.poly1d(poly_coeff)

x_interp = np.linspace(0, 5, 100)
y_spline = cs(x_interp)
y_poly = poly_fun(x_interp)

plt.figure(figsize=(10, 6))

plt.scatter(x, y, color='purple', label='Original Data Points', zorder=5)

plt.plot(x_interp, y_spline, label='Cubic Spline', linewidth=2)

plt.plot(x_interp, y_poly, label='Degree 5 Polynomial', linestyle='--', color='green')

plt.title('Cubic Spline vs Polynomial Fitting')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
