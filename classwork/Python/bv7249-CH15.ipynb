```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -2 * y

def analytical_sol(x):
    return np.exp(-2 * x)

x_start, x_end = 0, 2
h_values = np.arange(0.01, 1.05, 0.05)

plt.figure(figsize=(12, 8))

for h in h_values:
    num_steps = int((x_end - x_start) / h)
    
    x_vals = np.linspace(x_start, x_end, num_steps + 1)
    y_vals = np.zeros(num_steps + 1)
    
    y_vals[0] = 1

    for i in range(num_steps):
        y_vals[i + 1] = y_vals[i] + h * f(x_vals[i], y_vals[i])
    
    y_exact = analytical_sol(x_vals)
    pointwise_error = np.abs(y_exact - y_vals)
    total_accumulated_error = np.sum(pointwise_error)

    print(f"Step size (h): {h:.2f} | Total Accumulated Error: {total_accumulated_error:.4f}")
    
    plt.plot(x_vals, y_vals, label=f"h = {h:.2f}")

x_fine = np.linspace(x_start, x_end, 500)
plt.plot(x_fine, analytical_sol(x_fine), 'k--', linewidth=2, label="Exact Solution")


plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method Approximation with Varying Step Sizes")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

As the step size gets bigger, there is an upward trend with the error value.
