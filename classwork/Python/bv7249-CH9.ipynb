```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Part 1 - Slope: {m:.4f}, Intercept: {c:.4f}\n")

plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='red', label='Original Data Points')
plt.plot(X, m*X + c, color='blue', label=f'Best fit line: y = {m:.2f}x + {c:.2f}')
plt.title('Part 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()


np.random.seed(42)
X_rand = np.random.rand(10) * 10
Y_rand = np.random.rand(10) * 10

A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T

m_rand, c_rand = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0]
print(f"Part 2 - Slope: {m_rand:.4f}, Random Intercept: {c_rand:.4f}")

plt.figure(figsize=(8, 5))
plt.scatter(X_rand, Y_rand, color='green', label='Random Data Points')

x_line = np.linspace(min(X_rand), max(X_rand), 100)
plt.plot(x_line, m_rand*x_line + c_rand, color='orange', label=f'Best fit line: y = {m_rand:.2f}x + {c_rand:.2f}')
plt.title('Part 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()


true_m = 3.0
true_c = 2.0

X_own = np.arange(0, 1.01, 0.01)

noise = np.random.normal(0, 0.2, len(X_own)) 

Y_own = true_m * X_own + true_c + noise

A_own = np.vstack([X_own, np.ones(len(X_own))]).T
m_calc, c_calc = np.linalg.lstsq(A_own, Y_own, rcond=None)[0]

print(f"1st Part Equation: y = {true_m}x + {true_c}")
print(f"Part 3 Equation: y = {m_calc:.4f}x + {c_calc:.4f}")

plt.figure(figsize=(8, 5))
plt.scatter(X_own, Y_own, color='lightgray', label='Data with Perturbations', s=15)
plt.plot(X_own, true_m * X_own + true_c, color='green', linestyle='--', linewidth=2, label=f'Part 1 Line: y = {true_m}x + {true_c}')
plt.plot(X_own, m_calc * X_own + c_calc, color='red', linewidth=2, label=f'Part 3 Line: y = {m_calc:.2f}x + {c_calc:.2f}')
plt.title('Part 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

For part 2, since the data is randomized now, there is no distinct pattern/linear relationship to the arrangement of points. Therefore, the line of best fit does not match the dots at all.
