```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

#Below is the Monte Carlo Pi Function
def monte_carlo(num_samples):
    inside = 0
    for _ in range(num_samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <=1:
            inside +=1
            
    return 4*inside/num_samples

    #This function generates x and y points randomly in a 1x1 square,
    #and the fraction of the points inside this square can be used to approximate the value of pi
    
#Main code body, computes error and makes the samples
runs = 100
samples = 1000

absolute_error=[0]*runs
relative_error=[0]*runs

for i in range(runs):
    estimated_pi = monte_carlo(samples)
    absolute_error[i] = abs(np.pi - estimated_pi)
    relative_error[i] = abs((np.pi - estimated_pi)/np.pi)
        
#plotting the results
plt.figure()
plt.plot(range(1, runs+1), absolute_error, marker = 'o')
plt.xlabel("Runs")
plt.ylabel("Absolute value of error")
plt.title("Error of Monte Carlo using 100 runs and 1000 samples")
plt.show()
```
